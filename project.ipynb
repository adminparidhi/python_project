```python
# Bookstore Management System
class Book:
    def __init__(self, book_id, title, author, price, stock):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.price = price
        self.stock = stock

    def __str__(self):
        return f"ID: {self.book_id}, Title: {self.title}, Author: {self.author}, Price: ${self.price}, Stock: {self.stock}"


class Customer:
    def __init__(self, customer_id, name, email):
        self.customer_id = customer_id
        self.name = name
        self.email = email

    def __str__(self):
        return f"ID: {self.customer_id}, Name: {self.name}, Email: {self.email}"


class Bookstore:
    def __init__(self):
        self.books = []
        self.customers = []
        self.sales = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added to inventory.")

    def add_customer(self, customer):
        self.customers.append(customer)
        print(f"Customer '{customer.name}' added to records.")

    def sell_book(self, book_id, customer_id):
        book = next((b for b in self.books if b.book_id == book_id), None)
        customer = next((c for c in self.customers if c.customer_id == customer_id), None)

        if book and customer:
            if book.stock > 0:
                book.stock -= 1
                self.sales.append({"book": book.title, "customer": customer.name, "price": book.price})
                print(f"Book '{book.title}' sold to '{customer.name}'.")
            else:
                print(f"Book '{book.title}' is out of stock.")
        else:
            print("Invalid book or customer ID.")

    def display_books(self):
        print("\nBook Inventory:")
        for book in self.books:
            print(book)

    def display_customers(self):
        print("\nCustomer Records:")
        for customer in self.customers:
            print(customer)

    def display_sales(self):
        print("\nSales Records:")
        for sale in self.sales:
            print(f"Book: {sale['book']}, Customer: {sale['customer']}, Price: ${sale['price']}")


# Main Program
if __name__ == "__main__":
    bookstore = Bookstore()

    # Adding some books
    bookstore.add_book(Book(1, "The Great Gatsby", "F. Scott Fitzgerald", 10.99, 5))
    bookstore.add_book(Book(2, "1984", "George Orwell", 8.99, 3))
    bookstore.add_book(Book(3, "To Kill a Mockingbird", "Harper Lee", 12.99, 2))

    # Adding some customers
    bookstore.add_customer(Customer(1, "Alice Johnson", "alice34@gmail.com"))
    bookstore.add_customer(Customer(2, "Bob Smith", "bob788@gmail.com"))

    # Selling books
    bookstore.sell_book(1, 1)
    bookstore.sell_book(2, 2)
    bookstore.sell_book(1, 2)  # Selling another copy of The Great Gatsby

    # Displaying records
    bookstore.display_books()
    bookstore.display_customers()
    bookstore.display_sales()
    
```
